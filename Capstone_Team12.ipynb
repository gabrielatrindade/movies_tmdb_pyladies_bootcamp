```python
# os = operating system
import os
import pandas as pd
import matplotlib.pyplot as plt

print("Path at terminal when executing this file")
print(os.getcwd())
```

```python
path_movies = "dataset/movies.csv"
path_credits = "dataset/credits.csv"

movies = pd.read_csv(path_movies)
credits = pd.read_csv(path_credits)
```

```python
# within () you provide a number how many records will be displayed
movies.head(10)
```

```python
credits.head(10)
```

```python
merge = movies.merge(credits, on = ["movie_id"])
```

```python
merge.head(500)
```

```python
# The more vote_count are given on a movie the higher the chance of better vote_average. 
merge.plot(kind='scatter',x='vote_average',y='vote_count',color='black')
plt.show()
```
